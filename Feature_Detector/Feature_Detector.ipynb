```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter

# Load the dataset
df = pd.read_csv('/kaggle/input/imdb-top-1000-movies/data.csv')

# Question 1: What is the average rating of all movies in the dataset?
# TODO: Write the function to calculate and return the average rating of all movies.
def find_average_rating(df):
    avg_rating = df['averageRating'].mean()  # Fix the error by completing the code
    print(f"Question 1: The average rating of all movies is {avg_rating:.2f}")
    return avg_rating

# Question 2: Which movie has the highest rating?
# TODO: Write the function to find the movie with the highest rating and print its title and rating.
def find_highest_rated_movie(df):
    highest_rated = df.loc[df['averageRating'].idxmax()]  # Ensure this is properly finding the movie with the highest rating
    print(f"Question 2: The highest rated movie is '{highest_rated['title']}' with a rating of {highest_rated['averageRating']}")
    return highest_rated['title'], highest_rated['averageRating']

# Question 3: What are the most common genres in the dataset?
# TODO: Write the function to extract and count the most common genres.
def find_most_common_genres(df):
    all_genres = []  # Start with an empty list to store genres
    for genres in df['genres']:
        # TODO: Split genres by comma and add to the list (Fix the error here)
        all_genres.extend(genres.split(','))  # Fix this line by using extend() to add all genres
    genre_counts = Counter(all_genres)
    most_common = genre_counts.most_common(5)

    print("Question 3: The most common genres are:")
    for genre, count in most_common:
        print(f"  {genre}: {count} movies")

    return most_common

# Question 4: How many movies were released in each decade?
# TODO: Write the function to group the movies by decade and plot the results.
def count_movies_by_decade(df):
    # Create a decade column (Fix the error where this step might not be working)
    df['decade'] = (df['releaseYear'] // 10) * 10
    decade_counts = df['decade'].value_counts().sort_index()

    print("Question 4: Number of movies by decade:")
    for decade, count in decade_counts.items():
        print(f"  {decade}s: {count} movies")

    # Create a bar chart
    plt.figure(figsize=(10, 6))
    decade_counts.plot(kind='bar', color='skyblue')
    plt.title('Number of Movies Released by Decade')
    plt.xlabel('Decade')
    plt.ylabel('Number of Movies')
    plt.xticks(rotation=45)
    plt.grid(axis='y', alpha=0.3)
    plt.tight_layout()
    plt.show()

    return decade_counts

# Main function to run all analyses
def analyze_movie_dataset(df):
    print("===== BEGINNER-FRIENDLY MOVIE DATASET ANALYSIS =====\n")

    # Run all analyses
    avg_rating = find_average_rating(df)  # This function is ready
    print()

    highest_movie, highest_rating = find_highest_rated_movie(df)  # This function is ready
    print()

    common_genres = find_most_common_genres(df)  # Fix the error here in the function
    print()

    decade_counts = count_movies_by_decade(df)  # This function is ready
    print()

    # Compile results
    results = {
        "average_rating": avg_rating,
        "highest_rated_movie": {"title": highest_movie, "rating": highest_rating},
        "common_genres": common_genres,
        "movies_by_decade": decade_counts,
    }

    return results

# Run the analysis
if __name__ == "__main__":  # Fix this if condition (should be __name__)
    analyze_movie_dataset(df)

```
